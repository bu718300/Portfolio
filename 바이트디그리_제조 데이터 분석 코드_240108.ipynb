````python
# íŒŒì¼ ë¡œë“œ
import pandas as pd
df = pd.read_csv('wafer.csv')

# Classê°€ NULLì¸ ë°ì´í„° ì œê±°
data_cleaned = df.dropna(subset=['Class'])

# ë°ì´í„° í™•ì¸
data_cleaned.head()
````

````python
from sklearn.preprocessing import StandardScaler

# íŠ¹ì„±ê³¼ ë ˆì´ë¸” ë¶„ë¦¬
X = data_cleaned.drop('Class', axis=1)
y = data_cleaned['Class']

# í‘œì¤€í™” ìŠ¤ì¼€ì¼ëŸ¬ ì´ˆê¸°í™” ë° ì ìš©
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ìŠ¤ì¼€ì¼ë§ëœ ë°ì´í„° í”„ë ˆì„ ìƒì„±
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# ê²°ê³¼ í™•ì¸
X_scaled_df.head()
````

````python
from sklearn.model_selection import train_test_split

# ë°ì´í„°ë¥¼ í•™ìŠµ ì„¸íŠ¸ì™€ í…ŒìŠ¤íŠ¸ ì„¸íŠ¸ë¡œ ë¶„ë¦¬
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.2, random_state=42)

# ë¶„ë¦¬ ê²°ê³¼ í™•ì¸
(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
````

````python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report

# ëœë¤í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# ê·¸ë˜ë””ì–¸íŠ¸ ë¶€ìŠ¤íŒ… ëª¨ë¸
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(X_train, y_train)

# ëœë¤í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ í‰ê°€
rf_predictions = rf_model.predict(X_test)
rf_report = classification_report(y_test, rf_predictions)

# ê·¸ë˜ë””ì–¸íŠ¸ ë¶€ìŠ¤íŒ… ëª¨ë¸ í‰ê°€
gb_predictions = gb_model.predict(X_test)
gb_report = classification_report(y_test, gb_predictions)

(rf_report, gb_report)
````

````python
import numpy as np

# ëœë¤í¬ë ˆìŠ¤íŠ¸ì™€ ê·¸ë˜ë””ì–¸íŠ¸ ë¶€ìŠ¤íŒ… ëª¨ë¸ì˜ ì˜ˆì¸¡ ê²°ê³¼ ê²°í•© (í‰ê· )
ensemble_predictions = np.round((rf_model.predict_proba(X_test)[:, 1] + gb_model.predict_proba(X_test)[:, 1]) / 2)

# ì•™ìƒë¸” ëª¨ë¸ í‰ê°€
ensemble_report = classification_report(y_test, ensemble_predictions)

ensemble_report
````

````python
from sklearn.metrics import precision_recall_curve

# ì•™ìƒë¸” ëª¨ë¸ì˜ í™•ë¥  ì˜ˆì¸¡
ensemble_probabilities = (rf_model.predict_proba(X_test)[:, 1] + gb_model.predict_proba(X_test)[:, 1]) / 2

# ì •ë°€ë„-ì¬í˜„ìœ¨ ê³¡ì„  ë° ì„ê³„ê°’ ê³„ì‚°
precisions, recalls, thresholds = precision_recall_curve(y_test, ensemble_probabilities)

# ì„ê³„ê°’ ì¡°ì •ì— ë”°ë¥¸ ì •ë°€ë„, ì¬í˜„ìœ¨ í™•ì¸
threshold_results = pd.DataFrame({
    'Threshold': thresholds,
    'Precision': precisions[:-1],
    'Recall': recalls[:-1]
})

threshold_results.head()
````

````python
# ëª‡ ê°€ì§€ ì„ê³„ê°’ì— ëŒ€í•œ ì„±ëŠ¥ ë¹„êµ
selected_thresholds = [0.1, 0.15, 0.2, 0.3, 0.5]  # ì„ê³„ê°’ ì˜ˆì‹œ
performance_comparison = []

for threshold in selected_thresholds:
    # ì„ê³„ê°’ì— ë”°ë¥¸ ì˜ˆì¸¡
    adjusted_predictions = (ensemble_probabilities >= threshold).astype(int)

    # ì„±ëŠ¥ í‰ê°€
    adjusted_report = classification_report(y_test, adjusted_predictions, output_dict=True)
    performance_comparison.append({
        'Threshold': threshold,
        'Precision (Class 1)': adjusted_report['1.0']['precision'],
        'Recall (Class 1)': adjusted_report['1.0']['recall'],
        'F1-Score (Class 1)': adjusted_report['1.0']['f1-score'],
        'Accuracy': adjusted_report['accuracy']
    })

# ê²°ê³¼ë¥¼ ë°ì´í„° í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜
performance_df = pd.DataFrame(performance_comparison)
performance_df
````

````python
# ì´ìƒì¹˜ì— ëŒ€í•œ ì˜¤íƒ(ì •ìƒì„ ì´ìƒìœ¼ë¡œ íŒë‹¨í•˜ëŠ” case)ì´ ë°œìƒí•˜ë”ë¼ë„
# ë¯¸íƒ(ì´ìƒì„ ë†“ì¹˜ëŠ” case)ì„ ì¤„ì´ëŠ”ê²Œ í›¨ì”¬ ì¤‘ìš”í•˜ë‹¤ ìƒê°í•˜ì—¬ ì„ê³„ì¹˜ë¥¼ 0.15ë¡œ ë‚®ì¶”ì—ˆìŠµë‹ˆë‹¤.
# (0.01ê¹Œì§€ë„ ë‚®ì¶°ë³´ë©° í…ŒìŠ¤íŠ¸ë¥¼ í•´ë³´ì•˜ìœ¼ë‚˜ ë¯¸íƒì˜ ë³€í™”ëŸ‰ì´ ì ì€ë° ë°˜í•´
# Class 0(ì •ìƒ)ì— ëŒ€í•œ ì •í™•ë„ê°€ ê¸°í•˜ê¸‰ìˆ˜ì ìœ¼ë¡œ ë‚®ì•„ì ¸ ì´ì— ëŒ€í•œ ë…¼ì˜ê°€ í•„ìš”í•©ë‹ˆë‹¤.)

# ì„ê³„ê°’ì„ 0.15ë¡œ ì„¤ì •í•˜ì—¬ ì˜ˆì¸¡ ìˆ˜í–‰
adjusted_threshold = 0.15
final_predictions = (ensemble_probabilities >= adjusted_threshold).astype(int)

# ìµœì¢… ëª¨ë¸ í‰ê°€
final_report = classification_report(y_test, final_predictions)

final_report
````

````python
import matplotlib.pyplot as plt
import numpy as np
import itertools
from sklearn.metrics import confusion_matrix, roc_curve, auc

# í˜¼ë™ í–‰ë ¬ ë° ROC ê³¡ì„ ì„ ë‹¤ì‹œ ê³„ì‚°í•´ì•¼ í•©ë‹ˆë‹¤ (ì½”ë“œ ì‹¤í–‰ ìƒíƒœê°€ ì´ˆê¸°í™”ë˜ì—ˆê¸° ë•Œë¬¸ì—)
# ë°ì´í„° ë° ëª¨ë¸ì˜ ì˜ˆì¸¡ ê²°ê³¼ë¥¼ ë‹¤ì‹œ ë¡œë“œí•˜ê±°ë‚˜ ê³„ì‚°í•˜ëŠ” ì½”ë“œë¥¼ ì¶”ê°€í•˜ê² ìŠµë‹ˆë‹¤.

# í˜¼ë™ í–‰ë ¬ ìƒì„±
conf_matrix = confusion_matrix(y_test, final_predictions)

# ROC ê³¡ì„ ê³¼ AUC ê³„ì‚°
fpr, tpr, _ = roc_curve(y_test, ensemble_probabilities)
roc_auc = auc(fpr, tpr)

# í˜¼ë™ í–‰ë ¬ ì‹œê°í™”
plt.figure(figsize=(10, 5))

# í˜¼ë™ í–‰ë ¬ subplot
plt.subplot(1, 2, 1)
plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(2)
plt.xticks(tick_marks, ['Class 0', 'Class 1'], rotation=45)
plt.yticks(tick_marks, ['Class 0', 'Class 1'])
plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')

# í˜¼ë™ í–‰ë ¬ì˜ ê° ì„¹ì…˜ì— ìˆ«ì í‘œì‹œ
thresh = conf_matrix.max() / 2.
for i, j in itertools.product(range(conf_matrix.shape[0]), range(conf_matrix.shape[1])):
    plt.text(j, i, format(conf_matrix[i, j], 'd'),
             horizontalalignment="center",
             color="white" if conf_matrix[i, j] > thresh else "black")

# ROC ê³¡ì„  subplot
plt.subplot(1, 2, 2)
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")

plt.show()
````
